```python
import torch.nn as nn
from config import config

class MultiTaskHeads(nn.Module):
    def __init__(self, hidden=768):
        super().__init__()

        self.disease = nn.Linear(hidden, config.NUM_DISEASE)
        self.advisory = nn.Linear(hidden, config.NUM_ADVISORY)
        self.entity = nn.Linear(hidden, config.NUM_ENTITY)

    def forward(self, x):
        return self.disease(x), self.advisory(x), self.entity(x)
```
