```python
import torch
from config import config
from models.full_model import AgroModel
from training.losses import MultiTaskLoss

class Trainer:
    def __init__(self):
        self.model = AgroModel().to(config.DEVICE)
        self.optimizer = torch.optim.AdamW(self.model.parameters(), lr=config.LR)
        self.loss_fn = MultiTaskLoss()

    def train_step(self, batch):
        input_ids, mask, labels = batch

        self.optimizer.zero_grad()
        d,a,e = self.model(input_ids, mask)

        loss = self.loss_fn.compute(d,a,e,*labels)
        loss.backward()
        self.optimizer.step()

        return loss.item()
```
