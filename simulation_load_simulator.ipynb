```python
import threading
import requests

def simulate_users(n=100):
    def hit():
        requests.post("http://localhost:8000/predict", json={"text": "tomato leaf disease"})

    threads = []
    for _ in range(n):
        t = threading.Thread(target=hit)
        t.start()
        threads.append(t)

    for t in threads:
        t.join()
```
