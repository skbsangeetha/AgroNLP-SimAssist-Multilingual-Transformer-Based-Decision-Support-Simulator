```python
import torch.nn as nn

class MultiTaskLoss:
    def __init__(self):
        self.ce = nn.CrossEntropyLoss()

    def compute(self, d_pred, a_pred, e_pred, d_true, a_true, e_true):
        loss_d = self.ce(d_pred, d_true)
        loss_a = self.ce(a_pred, a_true)
        loss_e = self.ce(e_pred, e_true)

        return 0.4*loss_d + 0.3*loss_a + 0.3*loss_e
```
